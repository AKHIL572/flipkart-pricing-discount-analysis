```python
# ==========================================================
# Import Required Libraries
# ==========================================================

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

plt.rcParams["figure.figsize"] = (8,5)
plt.rcParams["figure.dpi"] = 120
```

```python
# ==========================================================
# Load Dataset
# ==========================================================

df = pd.read_csv("../data/flipkart_dataset.csv")

print("="*60)
print("Dataset Loaded Successfully")
print("="*60)

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

df.head()
```

# Data Understanding

Before cleaning the dataset, it is important to understand its overall structure, data types, descriptive statistics, missing values, and duplicate records.

This initial assessment helps identify potential data quality issues that may affect the reliability of the analysis.

```python
# ==========================================================
# Select Required Columns
# ==========================================================

df = df[
    [
        "brand",
        "product_name",
        "product_category_tree",
        "product_rating",
        "retail_price",
        "discounted_price",
    ]
]

print(f"Dataset Shape : {df.shape}")

display(df.head())

print("\nDataset Information\n")
display(df.info())

print("\nDescriptive Statistics\n")
display(df.describe())
```

```python
# ==========================================================
# Missing Value Analysis
# ==========================================================

missing = pd.DataFrame(
    {
        "Missing Values": df.isnull().sum(),
        "Missing Percentage": (
            df.isnull().sum() / len(df) * 100
        ).round(2),
    }
)

display(missing)
```

## Observations

The initial assessment of missing values reveals several important data quality issues.

- **Brand** contains **5,864 missing values (~29%)**, which is too large to remove without losing a significant portion of the dataset. These values will be replaced with **"Unknown"** to preserve the records.

- **Retail Price** and **Discounted Price** each contain **78 missing values (<1%)**. Since pricing is essential for this analysis, these rows will be removed.

- **Product Rating** currently shows no missing values because unavailable ratings are stored as the text **"No rating available"** rather than actual null values. This column will be converted to numeric later, after which these entries will become **NaN**.

```python
# ==========================================================
# Duplicate Analysis
# ==========================================================

exact_duplicates = df.duplicated().sum()

print(f"Exact Duplicate Rows : {exact_duplicates}")

soft_duplicates = df.duplicated(
    subset=[
        "brand",
        "product_name",
        "retail_price",
        "discounted_price",
    ],
    keep=False,
)

print(f"Potential Duplicate Listings : {soft_duplicates.sum()}")

display(
    df[soft_duplicates]
    .sort_values("product_name")
    .head(10)
)
```

## Duplicate Handling Decision

The dataset contains both exact duplicates and repeated product listings.

After inspecting the repeated records, the following strategy was adopted:

- Exact duplicate rows are removed because they contain identical information.
- Products having different retail or discounted prices are retained since they likely represent different product variants or listing updates.
- Duplicate removal is performed using **brand**, **product name**, **retail price**, and **discounted price**, ensuring that meaningful pricing variations are preserved while redundant records are removed.

```python
# ==========================================================
# Data Cleaning
# ==========================================================

# Convert rating column to numeric
df["product_rating"] = pd.to_numeric(
    df["product_rating"],
    errors="coerce",
)

# Convert prices to numeric
df["retail_price"] = pd.to_numeric(
    df["retail_price"],
    errors="coerce",
)

df["discounted_price"] = pd.to_numeric(
    df["discounted_price"],
    errors="coerce",
)

# Remove rows with missing prices
df = df[
    df["retail_price"].notna()
    & df["discounted_price"].notna()
]

# Fill missing brands
df["brand"] = df["brand"].fillna("Unknown")

# Remove duplicate listings
df = df.drop_duplicates(
    subset=[
        "brand",
        "product_name",
        "retail_price",
        "discounted_price",
    ]
)

print("Cleaning Completed Successfully.")

print(f"Final Dataset Shape : {df.shape}")

display(df.isnull().sum())
```

```python
# ==========================================================
# Feature Engineering
# ==========================================================

# Discount Percentage
df["discount_percentage"] = (
    (
        df["retail_price"]
        - df["discounted_price"]
    )
    / df["retail_price"]
) * 100

# Main Category
df["main_category"] = (
    df["product_category_tree"]
    .str.replace(r'[\[\]"]', "", regex=True)
    .str.split(">>")
    .str[0]
    .str.strip()
)

# Sub Category
df["sub_category"] = (
    df["product_category_tree"]
    .str.replace(r'[\[\]"]', "", regex=True)
    .str.split(">>")
    .str[1]
    .str.strip()
)

# Price Bucket
df["price_bucket"] = pd.cut(
    df["discounted_price"],
    bins=[0, 500, 1000, 2000, 5000, 600000],
    labels=[
        "<₹500",
        "₹500–1K",
        "₹1K–2K",
        "₹2K–5K",
        "₹5K+",
    ],
)

display(df.head())
```

```python
# ==========================================================
# Export Clean Dataset
# ==========================================================

df.to_csv(
    "../data/cleaned_flipkart_dataset.csv",
    index=False,
)

print("Cleaned dataset exported successfully.")

print(f"Final Rows    : {df.shape[0]}")
print(f"Final Columns : {df.shape[1]}")
```
